```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/cleaned_htic_data.csv')
```

```python
df
```

```python
cum_time=df['Cumulative time'].to_list()
arti_hr=df['Artifact corrected heart rate'].to_list()
```

```python
print(cum_time,arti_hr,sep="\n\n")
```

```python
import matplotlib.pyplot as plt
plt.plot(cum_time,arti_hr,linewidth=0.7)
# giving a title to my graph
plt.title('Heart rate Signal')
plt.xlabel('time')
plt.ylabel('heart rate')
plt.figure(figsize=(1,1))
plt.show()
```

```python
off=0
rms_noise=6
noise=np.random.normal(loc=off,scale=rms_noise,size=11576)
measured=np.std(noise)
print(measured)
```

```python
hrnoise=noise.tolist()
```

```python
plt.plot(cum_time,hrnoise,linewidth=0.7)
# giving a title to my graph
plt.title('Noise Signal')
plt.xlabel('time')
plt.ylabel('Noise')
plt.figure(figsize=(1,1))
plt.show()
```

```python
hr_arr=np.array(arti_hr)
no_arr=np.array(hrnoise)
print(hr_arr)
print(no_arr)
```

```python
sig=hr_arr + no_arr
print(sig)
```

```python
plt.plot(cum_time,sig,linewidth=0.7)
# giving a title to my graph
plt.title('Noise added Signal')
plt.xlabel('time')
plt.ylabel('HR')
plt.figure(figsize=(1,1))
plt.show()
```

```python
plt.title('Noise added Signal')
plt.xlabel('time')
plt.ylabel('HR')
plt.plot(cum_time, sig, label = "Noise added Signal",linewidth=0.4)
plt.plot(cum_time, arti_hr, label = "filtered Signal",linewidth=0.9)
plt.legend()
plt.show()
```
